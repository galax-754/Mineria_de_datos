<h3>Introducción al Análisis de Texto</h3>
<p>El análisis de texto es el proceso de extraer información significativa a partir de datos de texto. Se utiliza para entender tendencias, patrones o simplemente visualizar el contenido de manera más clara. Uno de los métodos más populares para visualizar palabras importantes en un conjunto de datos es mediante una <strong>nube de palabras</strong> (word cloud).</p>
<p>Una <strong>word cloud</strong> representa las palabras más frecuentes de un texto: las palabras que aparecen más veces se muestran con un tamaño más grande, mientras que las menos frecuentes se muestran más pequeñas. Esto ayuda a identificar rápidamente los temas principales o las palabras clave en los datos de texto.</p>
<p>En esta práctica, aprenderemos cómo generar una nube de palabras utilizando Python.</p>

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

<h3>2. Cargar el texto o los datos</h3>


```python
# Texto de ejemplo
texto = "Kareny. Carlos. Jania. Ernesto."
```

<h3>3. Crear la nube de palabras</h3>


```python
# Crear objeto WordCloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(texto)

# Mostrar la imagen
plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Nube de Palabras / Word Cloud', fontsize=20)
plt.show()
```
