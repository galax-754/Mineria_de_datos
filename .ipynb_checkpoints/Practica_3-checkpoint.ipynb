```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

<h3>Cargo el dataset</h3>

```python
df = pd.read_csv("csv/hour_cleaned.csv")
df.head()
```

<h3>Gráfico 1: Pie Chart (gráfico de pastel)</h3>

```python
# Contar cuántas veces aparece cada tipo de clima
clima_counts = df['weathersit'].value_counts()

# Dibujar gráfico de pastel
plt.figure(figsize=(6,6))
plt.pie(clima_counts, labels=clima_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribución de Clima')
plt.show()
```

<h3> Gráfico 2: Histogram (Histograma)</h3>

```python
plt.figure(figsize=(8,6))
plt.hist(df['temp'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribución de Temperatura')
plt.xlabel('Temperatura')
plt.ylabel('Frecuencia')
plt.show()
```

<h3>Gráfico 3: Boxplot (Diagrama de caja)</h3>

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='season', y='hum', data=df)
plt.title('Humedad por Estación')
plt.xlabel('Estación')
plt.ylabel('Humedad')
plt.show()
```

<h3>Gráfico 4: Line Plot (Gráfico de líneas)</h3>

```python
plt.figure(figsize=(12,6))
plt.plot(pd.to_datetime(df['dteday']), df['cnt'], color='green')
plt.title('Demanda de Bicicletas a lo largo del tiempo')
plt.xlabel('Fecha')
plt.ylabel('Cantidad de bicicletas rentadas')
plt.show()
```

<h3> Gráfico 5: Scatter Plot (Gráfico de dispersión)</h3>

```python
plt.figure(figsize=(8,6))
plt.scatter(df['temp'], df['cnt'], alpha=0.5)
plt.title('Relación entre Temperatura y Cantidad de Bicicletas Rentadas')
plt.xlabel('Temperatura')
plt.ylabel('Cantidad de bicicletas')
plt.show()
```

<h3>BONUS: For Loop para hacer muchos gráficos automáticamente</h3>

```python
numericas = ['temp', 'atemp', 'hum', 'windspeed', 'casual', 'registered', 'cnt']

for columna in numericas:
    plt.figure(figsize=(8,6))
    plt.hist(df[columna], bins=20, color='lightcoral', edgecolor='black')
    plt.title(f'Distribución de {columna}')
    plt.xlabel(columna)
    plt.ylabel('Frecuencia')
    plt.show()
```
