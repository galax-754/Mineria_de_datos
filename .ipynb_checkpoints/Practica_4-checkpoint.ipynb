<h1>Practica 4 Statistic Test</h1>

<p>Objetivo:
<li>Analizar cómo se relacionan las variables numéricas.</li>
    <li>Calcular y visualizar la matriz de correlación.</li>
 <p>

<h3>Importar liberias necesarias</h3>

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

<h3>Cargar el dataset</h3>

```python
df = pd.read_csv('csv/hour_cleaned.csv')
df.head()
```

<h3>Calcular la matriz de correlacion</h3>

```python
# Filtrar solo columnas numéricas
df_num = df.select_dtypes(include=['number'])
#calculamos la correlación
correlation_matrix = df_num.corr()
```

<h3>Matriz de correlacion</h3>

```python
plt.figure(figsize=(14,10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlaciones')
plt.show()
```

<p>
Rojo oscuro ➔ alta correlación positiva. <br>
Azul oscuro ➔ alta correlación negativa.</p>
