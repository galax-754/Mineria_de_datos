<h3>Importo librerias </h3>

<h3>Cargo el dataset y muestro las primeras filas</h3>

```python
df = pd.read_csv("csv/hour.csv")
df.head()
```

<h3>Revision de tamaño del dataset, tipos de datos, y nulos</h3>

```python
#Dimensiones
print("Shape:", df.shape)

#Tipos de datos 
df.dtypes

#Nulos
df.isnull().sum()
```

<h3>Columnas clave</h3>
<p>
    <li>dteday: Fecha</li>
    <li>temp, atemp, hum, windspeed: valores numéricos</li>
    <li>weathersit: valor alfanumérico (categoría)</li>
    <li>cnt: variable dependiente (total bicicletas)</li>
    <li>hr: hora del día</li>
    <li>weekday: día de la semana</li>
</p>

<h3>Conversion de tipos</h3>

```python
df['dteday'] = pd.to_datetime(df['dteday'])
```

```python
## Convertir columnas categóricas
categorical_cols = ['season', 'yr', 'mnth', 'hr', 'holiday', 'weekday', 'workingday', 'weathersit']
for col in categorical_cols:
    df[col] = df[col].astype('category')
```

```python
df.dtypes
```

<h3>Elimino columnas innecesarias</h3>

```python
df.drop(['instant'], axis=1, inplace=True)
```

<h3>Guardo el dataset limpio</h3>

```python
df.to_csv("hour_cleaned.csv", index=False)
```
